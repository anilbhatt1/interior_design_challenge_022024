```python
import numpy as np 
import matplotlib.pyplot as plt 
import glob 
import imageio
from PIL import Image
import cv2
```

```python
file_names = glob.glob('ikea_mask/*.jpg')
```

```python
for i, f in enumerate(file_names): 
    
    mask_orig = imageio.imread(f)
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (50,50))
    # mask_mod = cv2.morphologyEx(mask_orig, cv2.MORPH_OPEN, kernel, iterations=3)
    mask_mod = cv2.dilate(mask_orig, kernel, iterations=4)
    
    f_mod = f.replace('ikea_mask', 'ikea_mask_broaden')
    mask_mod = Image.fromarray(mask_mod)
    mask_mod.save(f_mod)
            
    # fig, axs = plt.subplots(1, 2, figsize=(10, 5))

    # axs[0].imshow(mask_orig)
    # axs[0].axis('off')
    
    # axs[1].imshow(mask_mod)
    # axs[1].axis('off')

    # # Display the plots
    # plt.tight_layout()
    # plt.show()
    
    # if i > 10: 
    #     break 
```
